```python
import pandas as pd
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)
print(df)
# dataframe in pandas is in the tabular form
```

```python
import pandas as pd
ages = pd.Series([22, 35, 58], name="Age")
print(ages)
print(ages.describe())
# describe ==> gather count, max, min, avg
```
