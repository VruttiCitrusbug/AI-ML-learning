```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
# The keras.Sequential function is used to stack layers linearly to create a neural network model. This means that the output of each layer feeds into the next layer in sequence.
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(1,), activation='relu'),
    keras.layers.Dense(1, activation='linear')
])
# First Layer:
# This line defines the first hidden layer of our neural network.
# Dense(10): It specifies a fully connected layer with 10 neurons. Each neuron will receive inputs from all neurons in the previous layer.
# input_shape=(1,): For the first layer, we need to specify the shape of the input data. Here, we have a single feature, so we use input_shape=(1,).
# activation='relu': The Rectified Linear Unit (ReLU) activation function is used in this layer. It introduces non-linearity, allowing the model to learn complex patterns in the data.
# Second Layer:
# This line defines the output layer of our neural network.
# Dense(1): It's a single neuron without specifying the input shape. This is because it automatically receives input from the previous layer.
# activation='linear': The linear activation function is used in the output layer. For regression tasks like predicting numerical values, a linear activation is commonly used as it allows the model to output a continuous range of values.
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
import numpy as np

# Generate random input data
X = np.random.rand(1000, 1)
# Generate corresponding output data (target)
y = X * 3 + 1 + np.random.randn(1000, 1) * 0.1

# Train the model
history = model.fit(X, y, epochs=50, verbose=0)
```

```python
# Plot the training loss over epochs
plt.plot(history.history['loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()

# Visualize the model's predictions
plt.scatter(X, y, label='Actual')
plt.scatter(X, model.predict(X), color='r', label='Predicted')
plt.legend()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import Adam

# Define your neural network model
model = Sequential()

# Add input layer (with input dimension)
model.add(Dense(64, input_dim=10))  # 10 input features, 64 neurons in the layer
model.add(Activation('relu'))  # Activation function: Rectified Linear Unit (ReLU)

# Add hidden layers
model.add(Dense(32))  # 32 neurons in the layer
model.add(Activation('relu'))

# Add dropout layer for regularization (optional)
model.add(Dropout(0.2))  # Dropout rate: 20%

# Add output layer
model.add(Dense(1))  # Output layer with 1 neuron for binary classification
model.add(Activation('sigmoid'))  # Activation function: Sigmoid for binary classification

# Compile the model
optimizer = Adam(learning_rate=0.001)  # Adam optimizer with learning rate 0.001
model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
# Loss function: Binary Crossentropy for binary classification
# Metrics to evaluate: Accuracy

# Print the model summary
model.summary()
```

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

# Load CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalize pixel values to between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0

# Define CNN architecture
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10)
])

# Compile the model
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Train the model
history = model.fit(train_images, train_labels, epochs=10, 
                    validation_data=(test_images, test_labels))

# Evaluate the model
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('\nTest accuracy:', test_acc)

# Plot training history
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.show()
```
