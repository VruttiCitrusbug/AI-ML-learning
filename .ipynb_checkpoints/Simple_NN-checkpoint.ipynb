```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
# The keras.Sequential function is used to stack layers linearly to create a neural network model. This means that the output of each layer feeds into the next layer in sequence.
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(1,), activation='relu'),
    keras.layers.Dense(1, activation='linear')
])
# First Layer:
# This line defines the first hidden layer of our neural network.
# Dense(10): It specifies a fully connected layer with 10 neurons. Each neuron will receive inputs from all neurons in the previous layer.
# input_shape=(1,): For the first layer, we need to specify the shape of the input data. Here, we have a single feature, so we use input_shape=(1,).
# activation='relu': The Rectified Linear Unit (ReLU) activation function is used in this layer. It introduces non-linearity, allowing the model to learn complex patterns in the data.
# Second Layer:
# This line defines the output layer of our neural network.
# Dense(1): It's a single neuron without specifying the input shape. This is because it automatically receives input from the previous layer.
# activation='linear': The linear activation function is used in the output layer. For regression tasks like predicting numerical values, a linear activation is commonly used as it allows the model to output a continuous range of values.
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
import numpy as np

# Generate random input data
X = np.random.rand(1000, 1)
# Generate corresponding output data (target)
y = X * 3 + 1 + np.random.randn(1000, 1) * 0.1

# Train the model
history = model.fit(X, y, epochs=50, verbose=0)
```

```python
# Plot the training loss over epochs
plt.plot(history.history['loss'])
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()

# Visualize the model's predictions
plt.scatter(X, y, label='Actual')
plt.scatter(X, model.predict(X), color='r', label='Predicted')
plt.legend()
plt.show()
```
