```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'x': [1, 2, 3, 4, 5],
    'y': [2, 4, 6, 8, 10]
}
df = pd.DataFrame(data)
```

```python
df.plot(x='x', y='y', kind='line')
plt.show()
```

```python
df.plot(x='x', y='y', kind='bar')
plt.show()
```

```python
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Title of the Plot')
plt.grid(True)
plt.legend(['Legend'])
plt.show()
```
